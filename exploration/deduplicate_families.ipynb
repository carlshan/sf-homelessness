```python
import pandas as pd
import clean

import networkx as nx
```

```python
program = clean.get_program()
```

```python
families = program[['Deduplicated Subject Unique Identifier', 'Family Site Identifier']].drop_duplicates().dropna().set_index(['Family Site Identifier'])
```

```python
families
```

```python
edges = families.merge(families, left_index=True, right_index=True)
```

```python
edges
```

```python
unique_edges = edges[edges['Deduplicated Subject Unique Identifier_x'] <= edges['Deduplicated Subject Unique Identifier_y']].drop_duplicates()
```

```python
unique_edges
```

```python
G = nx.Graph([tuple(e) for e in unique_edges.values])
```

```python
families = [f for f in nx.connected_components(G)]
```

```python
large_families = [f for f in families if len(f) > 5]
```

```python
print len(families)
print len(large_families)
print str(float(len(large_families))/len(families)*100) + "%"
```

```python
large_families
```

```python
columns = ['Subject Unique Identifier','Deduplicated Subject Unique Identifier','Converted DOB','Program Name','Converted Program Start Date','Converted Program End Date']
```

```python
program[program['Deduplicated Subject Unique Identifier'].isin(large_families[0])].set_index('Family Site Identifier').sort().sort('Converted Program Start Date')[columns]
```

- There are three densely connected graphs here (these are lists of Deduplicated SUIDs) :

- The reason they appear to be the same family is that we see the same Family Site ID assigned over different times to different families (i.e. FSID 6821)
