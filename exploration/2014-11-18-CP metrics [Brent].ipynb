Ike - 
Your proposed solution answers a different question: "How many unique 
families were ever in shelter in a given year?"  I think we need to 
clarify what is meant by "on average" here.  Given that, your proposed 
solution assumes that people don't stay in for more than a year, which 
doesn't generalize to other kinds of programs.  Instead, (e.g. 2013) 
look at the number of families who entered before January 1 2014 and 
exited after January 1 2013.    Also, similar to question 10, I think 
we should consider timeboxing based on when a family *entered*, not if 
they had any overlap at all

Methodology -

(1) Look at 2009 - 2013, and count number of unique families who entered in these given years

(2) Look at 2009 - 2013, and count number of unique who spent any time in any of these years

```python
# (1) 
# Examining for 2013
individuals_in_families = program[(program['Program Start Year'] == 2013)
                            & program['Family?']
                            & program['Program Name'].isin(valid_programs)]
```

```python
# 188 families entered in 2013
individuals_in_families['Family Identifier'].unique().shape
```

```python
shelter_programs = {name: {'families': [],
                           'individuals': []} for name in valid_programs}

for shelter in shelter_programs:
    num_unique_fams = []
    num_unique_inds = []
    for year in range(2009, 2014):
        individuals_in_families = program[(program['Program Start Year'] == year)
                                & program['Family?']
                                & program['Program Name'].isin([shelter])]
        num_fams = individuals_in_families['Family Identifier'].unique().shape[0]
        num_inds = individuals_in_families['Subject Unique Identifier'].unique().shape[0]
        num_unique_fams.append(num_fams)
        num_unique_inds.append(num_inds)
    shelter_programs[shelter]['families'] = num_unique_fams
    shelter_programs[shelter]['individuals'] = num_unique_inds
```

```python
plt.title('# Unique Homeless Families/Individuals \n Entering Shelter')
plt.plot([2009, 2010, 2011, 2012, 2013], num_unique_fams, label='# Unique Families')
plt.plot([2009, 2010, 2011, 2012, 2013], num_unique_inds, label='# Unique Inds')
plt.xticks([2009, 2010, 2011, 2012, 2013], [2009, 2010, 2011, 2012, 2013])
plt.ylabel('Count')
plt.xlabel('Year')
plt.legend(loc='best')
```

## Metric 7: How many families were in shelters on an average night in each quarter over the past four years (use as much data as is available for most recent quarter)?

```python
def is_between_program_start_end(row, date):
    program_start = row['Program Start Date']
    program_end = row['Program End Date']
    return is_between_dates(date, program_start, program_end)
```

```python
START_DATE = '1/1/2009'
END_DATE = '12/31/2013'
```

```python
time_series = pd.Series(index=pd.date_range(start=START_DATE, end=END_DATE, freq='D'))
```

```python
filtered_program = program[program['Program Name'].isin(valid_programs)
                               & program['Program End Date'].notnull()
                               & program['Family?']]
```

```python
for day in time_series.index:
    try: 
        f = filtered_program[filtered_program.apply(is_between_program_start_end, date=day,
                                         axis=1)]
        num_families = f['Family Identifier'].unique().shape[0]
        time_series[day] = num_families
    except Exception as e:
        print day
```
