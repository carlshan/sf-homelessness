## Metric 9
How long did the average family stay in a shelter in each quarter over the past four years? (could we show the distribution of time spent by families - are there some with very long stays)?

```python
def is_between_dates(date, begin, end):
    """
    Returns:
    ----------
    Returns a boolean indicating whether date is between begin and end
    
    Parameters: 
    ----------
    date : datetime object
    begin : datetime object
    end : datetime object
    """
    if pd.isnull(date):
        return False
    return date >= begin and date <= end
```

```python
# Testing that this function works as I think it should
begin = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2012, 3, 31)
```

```python
mask = program['Program Start Date'].apply(is_between_dates, 
                                begin=begin, 
                                end=end
```

```python
q1_2012 = program[  mask 
                  & program['Family?']
                  & program['Program End Date'].notnull()
                  & program['Program Name'].isin(valid_programs)]
```

```python
q1_2012.sort('Program Start Date')[columns].head()
```

```python
# Each family stay is a unique entry
groups = q1_2012.groupby(['Family Identifier', 'Program Start Date'])
```

```python
mn = groups['Days Stayed'].mean()
med = groups['Days Stayed'].median()
mx = groups['Days Stayed'].max()
```

```python
# It turns out, families stay together and don't have one individual leaving early
# than another. 
# I know this because the mean == median == max for each family stay
# which can only happen if there's uniform distribution for all members
# of a family during a stay
med.values == mx.values
```

```python
## Actually plotting 
```

```python
def get_days_stayed_for_quarters_in_year(year, program, valid_programs, 
                                         aggfunc=np.median):
    """
    Returns list of 4 values. Each value is the median (or another aggfunc) 
    of the days stayed by each family during the quarter.
    
    e.g., [56, 60, 70, 55] would indicate the median # of days stayed by
    families that entered in Q1 of the year is 56.
    60 for those who entered in Q2 of the year, and so forth.
    """
    
    quarter_values = []

    q1 = (datetime.datetime(year, 1, 1), datetime.datetime(year, 3, 31))
    q2 = (datetime.datetime(year, 4, 1), datetime.datetime(year, 6, 30))
    q3 = (datetime.datetime(year, 7, 1), datetime.datetime(year, 9, 30))
    q4 = (datetime.datetime(year, 10, 1), datetime.datetime(year, 12, 31))
    
    quarters = (q1, q2, q3, q4)
    
    for quarter in quarters:
        begin = quarter[0]
        end = quarter[1]
        mask = program['Program Start Date'].apply(is_between_dates, 
                                        begin=begin, 
                                        end=end)
        quarter = program[mask 
                          & program['Family?']
                          & program['Program End Date'].notnull()
                          & program['Program Name'].isin(valid_programs)]
        groups = quarter.groupby(['Family Identifier', 'Program Start Date'])
        days_per_stay = groups['Days Stayed'].mean() # taking mean is same as med/max/min
        val = aggfunc(days_per_stay)

        quarter_values.append(val)

    return quarter_values
```

```python
year_range = range(2009, 2014)
```

```python
vals = []
for year in year_range:
    q = get_days_stayed_for_quarters_in_year(year, program, valid_programs=valid_programs,
                                         aggfunc=np.median)
    vals += q
```

```python
plt.title('Median # Days Spent by Families \n (per quarter per year)')
plt.plot(range(1, len(vals)+1), vals)

# Labeling the xticks Q1 '9, Q2 '09 and so forth.
tick_labels = []
for year in year_range:
    tick_labels.append('Q1 \'{}'.format(str(year)[-2:]))
plt.xticks(range(1, len(med)+1,4 ), tick_labels)
```
