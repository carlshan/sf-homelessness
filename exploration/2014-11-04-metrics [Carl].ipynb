## Metric 1: How many families are there currently in shelter?

```python
# Someone is still in a shelther if they entered a program, but 
# does not have a recorded exit date
all_individuals_still_in_shelter = program[program['Program Start Date'].notnull() 
                                           & program['Program End Date'].isnull()]
```

```python
all_individuals_still_in_shelter[columns].head()
```

```python
all_individuals_in_families = all_individuals_still_in_shelter[all_individuals_still_in_shelter['Family?']]
```

```python
all_individuals_in_families.head()
```

```python
all_families = all_individuals_in_families['Family Identifier'].unique()
```

```python
# How many families are there in shelters?
len(all_families)
```
