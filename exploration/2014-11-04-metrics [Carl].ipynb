```python
program['Days Stayed'] = program['Stay Duration'].apply(lambda t: t.astype('timedelta64[D]')/np.timedelta64(1, 'D'))
```

```python
def normalize_days_stayed(days):
    if pd.isnull(days):
        return days
    elif days >= 90:
        return 90
    else:
        return days
```

```python
program['Normalized Days Stayed'] = program['Days Stayed'].apply(normalize_days_stayed)
```

## Size of families

```python
year = 2009
```

```python
point_in_time = datetime.datetime(year, 1, 15)
```

```python
pit_counts = program.loc[program['Program Start Date'].notnull()
            & program['Program End Date'].notnull()
            & program['Program Name'].isin(valid_programs)
            & program['Family?']
            & (program['Program Start Date'] <= point_in_time)
            & (program['Program End Date'] >= point_in_time)].copy()
```

```python
def get_family_size(row, df):
    family_id = row['Family Identifier']
    results = df[df['Family Identifier'] == family_id]
    num_individuals = results['Subject Unique Identifier'].unique().shape[0]
    return num_individuals
```

```python
pit_counts['Subject Unique Identifier'].unique().shape[0]
```

```python
pit_counts['Family Size'] = pit_counts.apply(get_family_size,
                 df=pit_counts,
                 axis=1)
```
