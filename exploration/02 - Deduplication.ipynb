# Deduplicating the HMIS

```python
import pandas as pd
```

```python
hmis_duplicates = pd.read_csv('../data/connecting_point/hmis_client_duplicates.csv')
program_data = pd.read_csv('../data/program with family.csv')
```

```python
hmis_duplicates.shape
```

```python
hmis_duplicates.columns
```

```python
merged_data = program_data.merge(hmis_duplicates, 
                   on = 'Subject Unique Identifier', 
                   how = 'inner')
```

```python
# TODO: For some reason, 21 individuals are not in this file 
# but were in HMIS duplicates.
# Question added for Candace to figure out why, and we'll come back to this
merged_data['Subject Unique Identifier'].unique().shape
```

```python
merged_data['Deduplicated Subject Unique Identifier'] = merged_data['Duplicate ClientID'].fillna(merged_data['Subject Unique Identifier'])
```

```python
# This looks liek what we expect, which is good!
merged_data[merged_data['Duplicate ClientID'].notnull()]
```

```python
merged_data.to_csv('../data/cleaned_data/deduplicated_program.csv',
                   index = False)
```

```python
unique_programs = grouped.agg({'Program Unique Identifier': lambda col: len(set(col[col.notnull()].values))})
```
