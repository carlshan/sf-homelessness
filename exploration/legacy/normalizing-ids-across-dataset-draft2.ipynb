```python
import pandas as pd
import clean
import networkx as nx
```

```python
matching = pd.read_csv('../data/matching/cp_hmis_match_results.csv').dropna()
```

```python
matching.columns = ['Raw Clientid', 'Raw Subject Unique Identifier', 'Found match?', 'match score']
```

```python
program = clean.get_program()
client = clean.get_cp_client()
```

```python
# merge
```

```python
matching = matching.merge(client[['Clientid','Raw Clientid']], how='left')
matching = matching.merge(program[['Subject Unique Identifier','Raw Subject Unique Identifier']], how='left')
```

```python
matching = matching.dropna(subset=['Clientid','Subject Unique Identifier'])
```

# Easy approach

```python
# NOTE: this approach assumes that we have no pathological cases (as follows), then checks to make sure everything is okay.
#
# Pathological case: Consider Connecting Point Clientids 1, 2 and HMIS SUIDs A, B, C.  If 1 is matched to A & B and
# 2 is matched to B & C, then this may result in A & B collapsing and 1 & 2 collapsing, but C remaining uncollapsed with
# A & B, (or some permutation of that).  Try drawing it out if it doesn't make sense.
#
# So, we check after to make sure that the entire table is properly matched one-to-one.
```

```python
# assign new SUID / Clientid based on multiple people from the *other* dataset matching to the same person
matching['New Clientid'] = matching.groupby('Subject Unique Identifier')['Clientid'].transform(min)
matching['New Subject Unique Identifier'] = matching.groupby('Clientid')['Subject Unique Identifier'].transform(min)
```

```python
matching_check = matching.drop_duplicates(['New Subject Unique Identifier','New Clientid'])
```

```python
# assertions
```

```python
len(matching_check['New Subject Unique Identifier'].unique()) == len(matching_check)
```

```python
len(matching_check['New Clientid'].unique()) == len(matching_check)
```

```python
# merge into program and client to get complete listing of all clients across dataset
```

```python
matching = matching[['Clientid','Subject Unique Identifier','New Clientid','New Subject Unique Identifier']].drop_duplicates()
```

```python
m_columns = ['Subject Unique Identifier','New Subject Unique Identifier','New Clientid','Clientid']
final_matching = matching.merge(program,
               on='Subject Unique Identifier',
               how='outer').merge(client,
                                  on='Clientid',
                                  how='outer')[m_columns].drop_duplicates()
```
