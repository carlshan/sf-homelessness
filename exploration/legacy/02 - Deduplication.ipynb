## Deduping the CP Data

```python
cp_duplicates = pd.read_csv('../data/connecting_point/cp_client_duplicates.csv')
cp_program_data = pd.read_csv('../data/connecting_point/client.csv')
```

```python
cp_merged_data = cp_program_data.merge(cp_duplicates[['clientid', 'Duplicate ClientID']],
                                       how='inner',
                                       left_on='Clientid',
                                       right_on='clientid')
```

```python
cp_merged_data['Deduplicated ClientID'] = cp_merged_data['Duplicate ClientID'].fillna(cp_merged_data['Clientid'])
```

```python
cp_merged_data['Deduplicated ClientID'].count()
```

```python
cp_merged_data.drop(['clientid', 'Duplicate ClientID'], axis=1, inplace=True)
```

```python
cp_merged_data.to_csv('../data/cleaned_data/connecting_point/deduped_client.csv')
```
