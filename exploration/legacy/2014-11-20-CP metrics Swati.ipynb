## Stacked chart showing number of families entering waitlist and duration of their stay (Written by Swati, please check)

```python
# Program to check if the servstart date is within beginning and end of a quarter
def is_between_dates(date, begin, end):
    """
    Returns:
    ----------
    Returns a boolean indicating whether date is between begin and end
    
    Parameters: 
    ----------
    date : datetime object
    begin : datetime object
    end : datetime object
    """
    if pd.isnull(date):
        return False
    return date >= begin and date <= end

## Define all the quarters


def quarter_analysis(program):
    months1=[] # no. of Families spending less than month on waitlist
    months2=[] # no. of FAmilies spending 1-2 months on waitlist
    months3=[] # no. of Families spending 2-3 months on waitlist
    months4=[] # no. of Families spending more than 3 months on the waitlist
    mean_days_spentbyfamily=[]
    min_days_spentbyfamily=[]
    max_days_spentbyfamily=[]
    Qbegin=[datetime.datetime(2012, 10, 1),datetime.datetime(2013, 1, 1),datetime.datetime(2013, 4, 1),
           datetime.datetime(2013, 7, 1),datetime.datetime(2013, 10, 1),datetime.datetime(2014, 1, 1),
           datetime.datetime(2014, 4, 1),datetime.datetime(2014, 7, 1)
           ]
    Qend=[datetime.datetime(2012, 12, 31),datetime.datetime(2013, 3, 31),datetime.datetime(2013, 6, 30),
           datetime.datetime(2013, 9, 30),datetime.datetime(2013, 12, 31),datetime.datetime(2014, 3, 31),
          datetime.datetime(2014, 6, 30),datetime.datetime(2014, 9, 30)
          ]  
    for i in range(len(Qbegin)):
        begin = Qbegin[i]
        end = Qend[i]
        mask = program['servstart'].apply(is_between_dates, 
                                        begin=begin, 
                                        end=end)
        quarter = program[mask 
                          & program['Family?']
                          & program['servend'].notnull()
                          & program['Days Stayed'].notnull()]
        groups = quarter.groupby(['Familyid','servstart'])
        
        days_per_stay = groups['Days Stayed'].mean()
        B=groups['Days Stayed'].mean()
        mean_days_spentbyfamily.append(B[B>0].mean())
        min_days_spentbyfamily.append(B[B>0].min())
        max_days_spentbyfamily.append(B[B>0].max())
        months1.append(B[(0 < B) & (B < 30)].size) # Families spending less than month on waitlist
        months2.append(B[(0 < B) & (B < 60)].size) # FAmilies spending 1-2 months on waitlist
        months3.append(B[(0 < B) & (B < 90)].size) # Families spending 2-3 months on waitlist
        months4.append(B[0 < B].size) # Families spending more than 3 months on the waitlist
    return months1,months2,months3,months4,mean_days_spentbyfamily,min_days_spentbyfamily,max_days_spentbyfamily
months1,months2,months3,months4,mean_days_spentbyfamily,min_days_spentbyfamily,max_days_spentbyfamily=quarter_analysis(program)
```
